```python
from __future__ import print_function
 
# Ignore warnings dari gensim
import warnings
warnings.filterwarnings(action='ignore', category=UserWarning, module='gensim')
 
import logging
import os.path
import sys
 
from gensim.corpora import WikiCorpus
```

```python
program = os.path.basename(sys.argv[0])
logger = logging.getLogger(program)
 
logging.basicConfig(format='%(asctime)s: %(levelname)s: %(message)s')
logging.root.setLevel(level=logging.INFO)
logger.info("running %s" % ' '.join(sys.argv))
 
namaFileInput = "./assets/idwiki-latest-pages-articles.xml.bz2"
namaFileOutput = "./assets/wiki_id.txt"
 
space = " "
i = 0
 
# Write file ke variabel namaFileOutput encoder utf-8
output = open(namaFileOutput, 'w', encoding='utf-8')
wiki = WikiCorpus(namaFileInput, dictionary={}, lower=True)
for text in wiki.get_texts():
    output.write(' '.join(text) + '\n')
    i = i + 1
    if i % 10000 == 0:
        logger.info("Saved " + str(i) + " articles")
 
output.close()
logger.info("Finished Saved " + str(i) + " articles")
```

### Train Embedding

```python
import time
import multiprocessing
from datetime import timedelta

from gensim.models import word2vec
start_time = time.time()
print('Creating Word2Vec Model...')
sentences = word2vec.LineSentence('./assets/wiki_id.txt')
word2vec_model = word2vec.Word2Vec(sentences, vector_size=300, workers=multiprocessing.cpu_count() -1)
word2vec_model.wv.save_word2vec_format('./assets/embedding.model', binary=False)
finish_time = time.time()

print('Finished. Elapsed time: {}'.format(timedelta(seconds=finish_time-start_time)))
```
